```python
import re
def findTotalWord(w):
    return re.compile(r'\b({0})\b'.format(w)).search    
```

```python
# incrocio id_pol_2021 con localizzaione-id --> stato(SIGLA)_pol  UNIVERLASE(mondo/usa)
stato_pol={}

for st in l_allID:
    pol_2=[]
    for st_1 in stati:
        if(findTotalWord(st_1)(st)!=None): 
            print(st_1+' -->in--> '+st)
            for i_2,p in id_pol.items():
                confronto_con = l_allID.get(st)
                for c in confronto_con:
                    if(c==i_2):
                        print(c+' == '+i_2)
                        print(p)
                        pol_2.append(p)
                        stato_pol[st]=pol_2
```

```python
#GET mondo: sigla_stato- lista_polarizzazione
get = {} 
for n in s:
    polar=[]
    for sp in stato_pol:
        if(findTotalWord(n)(sp)!=None):
            for j in stato_pol.get(sp):
                polar.append(j)
    get[n] = polar
```

```python
#END mondo
end={}

for g in get:
    if(get.get(g) !=[]):
        end[g]=get.get(g)
```

```python
# GET STATI AMERICA-> id_pol_2019 con localizzaione-id --> stato pol 
stato_pol_america={}
for st in l_allID:
    pol_america=[]
    for a in stati_america_elenco:    
        if(findTotalWord(a)(st)!=None):  
            print(a+' -->in--> '+st)
            for i_2,p in id_pol.items():
                confronto_con = l_allID.get(st)
                for c in confronto_con:
                    if(c==i_2):
                        print(c+' == '+i_2)
                        print(p)
                        pol_america.append(p)
                        stato_pol_america[st]=pol_america
```

```python
#get per america
get_america={}
for a in stati_america_elenco:       
    polar_america=[]           
    for sp in stato_pol_america:
        if(findTotalWord(a)(sp)!=None): 
            for k in stato_pol_america.get(sp):
                polar_america.append(k)
    get_america[a] = polar_america
```

```python
#CREA TOPICS-> PER OGNI STATO PRO, CONTRO E NEUTRAL
stati=[]
sentiment=[]
for e,u in end.items():
    for n in u:
        if(n>0.0):
            stati.append(e)
            sentiment.append('pro')
        elif(n==0.0):
            stati.append(e)
            sentiment.append('neutral')
        elif(n<0.0):
            stati.append(e)
            sentiment.append('contro')
```

```python
#get per america
get_sigla_america={}
for a in stati_america_elenco:
    id_tweet=[]           
    for l in l_allID:
        if(findTotalWord(a)(l)!=None): 
            for k in l_allID.get(l):
                id_tweet.append(k)
    get_sigla_america[a] = id_tweet
```

```python
end_sigla_usa={}
for g in get_sigla_america:
    if(get_sigla_america.get(g)!=[]):
        end_sigla_usa[g]=get_sigla_america.get(g)
```

creato dizionario per prendere chiave ID TWEET( valori biden e trump)

```python
with open('tweet_clean2021.json', 'r') as f:
        id_tweet=json.load(f)
```

```python
with open('id_pol_2021.json', 'r') as f:
        id_pol=json.load(f)
```

```python
id_presidente={}


for i,t in id_tweet.items():
    if('biden' in t) or ('BIDEN' in t) or ('Biden' in t):
        for d,p in id_pol.items():   
            if(p>=0):
                id_presidente[i]='biden'
            else:
                id_presidente[i]='trump'
    if('trump' in t) or ('TRUMP' in t) or ('Trump' in t):
            for d,p in id_pol.items():   
                if(p>=0):
                    id_presidente[i]='trump'
                else:
                    id_presidente[i]='biden'
```

```python
sigla=[]
presidente=[]
for s,l in end_sigla_usa.items():
    for r,t in regioni.items():
        if(s in t):
            for d,p in id_presidente.items():
                if(d in l):
                    sigla.append(r)
                    presidente.append(p)
```

```python
#CREA DF PER GRAFICO regioni usa
df_regioni_presidente = {'REGIONI':  sigla, 'presidente': presidente }
df_regioni_presidente = pd.DataFrame (df_regioni_presidente, columns = ['REGIONI','presidente'])
print (df_regioni_presidente)
```

```python
sns.set_theme(style='whitegrid')
sns.catplot(x="REGIONI", hue="presidente", kind="count", palette="pastel",edgecolor=".6",data=df_regioni_presidente, height=6.7, aspect=10.7/7.27)
```
